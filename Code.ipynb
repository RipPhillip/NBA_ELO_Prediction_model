# Clean up the data

```python
#Dropping basic stats
clean_df.drop(df.columns[0:2], axis=1, inplace=True)
```

```python
# Drop all basic stats
clean_df.drop(df.columns[6:22], axis=1, inplace=True)
```

```python
# Drop the random columns
clean_df.drop(clean_df.columns[[16, 21]], axis=1, inplace=True)
```

```python
# Rename the columns
clean_df.rename(columns={'\xa0':'Home', 'Opp.1':'Opp_pts', 'eFG%.1':'Opp eFG%', \
                         'TOV%.1':'Opp TOV%','FT/FGA.1':'Opp FT/FGA', 'W/L':'Win'}, inplace=True)
```

```python
# Fill the NaN values in the Home column
clean_df['Home'] = clean_df['Home'].fillna(1)
```

```python
# Drop those 0 values
clean_df.dropna(axis=0, inplace=True)
```

```python
# Make Home column a 1 and 0 for values
clean_df.replace(to_replace='@', value=0, inplace=True)
```

```python
# Make the W/L column 1 and 0, and an integer
clean_df.replace(to_replace={'Win':{'W':1, 'L':0}}, inplace=True)
```

```python
display(clean_df.head())
clean_df.info()
```

```python
clean_df['H_team_pts'] = np.where(clean_df['Home']==1, clean_df['Tm'], clean_df['Opp_pts'])
```

```python
clean_df['A_team_pts'] = np.where(clean_df['Home']==0, clean_df['Tm'], clean_df['Opp_pts'])
```
