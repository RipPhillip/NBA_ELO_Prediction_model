```python
years = [str(i) for i in range(2020, 2021)]
teams = ["ATL", "BRK", "BOS", "CHO", "CHI", "CLE", "DAL", "DEN",
                "DET", "GSW", "HOU", "IND", "LAC", "LAL", "MEM", "MIA",
                "MIL", "MIN", "NOP", "NYK", "OKC", "ORL", "PHI", "PHO",
                "POR", "SAC", "SAS", "TOR", "UTA", "WAS", "NJN", "NOH", "CHA", "CHH", "SEA", "VAN"
                "NOK"]
```

## New web scrape code

```python
# pseudocode:    
# create a master dataframe(initially empty) with the headers

# this is the HTML from the given URL
html = requests.get(url, proxies=proxy)
soup = BeautifulSoup(html.text)

# use getText() to extract the text of the adv. table header (stat/column names)
adv_headers = [th.getText() for th in soup.findAll('thead', limit=2)[0].findAll('th')]
# keep desired stats/column names
adv_headers = adv_headers[9:]

# initialize master_df, will have header with no rows
master_df = pd.DataFrame(columns = (adv_headers + ['Team', 'Season']) )

# for each year (2010 - 2019, 10 seasons), get the teams' gamelogs
```

```python
# for each team, get the basic and advanced game logs
for year in years:
############################## Teams Full Game Logs 2014 ################################
    for team in teams:
        # for the basic game log, remove opponent stats (by indexing)
        # NBA season we will be analyzing
        
        # URL where adv. game log table is
        url = "https://www.basketball-reference.com/teams/{}/{}/gamelog-advanced/".format(team, year)

        # this is the HTML from the given URL
        html = requests.get(url, proxies=proxy)
        soup = BeautifulSoup(html.text)

        # begin at third table row (first game log)
        rows = soup.findAll('tr')[2:]
        # extract stats of desired columns. nested list of each game log (row)
        adv_gl_lists = [[td.getText() for td in rows[i].findAll('td')]
                        for i in range(len(rows))]
        # create df for adv. game logs
        df_adv = pd.DataFrame(adv_gl_lists, columns = adv_headers)
        
        # Add the team names to the DF
        df_adv['Team'] = team
        
        # Add the season year to the DF
        df_adv['Season'] = year

        # add the dataframe to the master dataframe
        master_df_11 = pd.concat([df_adv, master_df])
        
```
